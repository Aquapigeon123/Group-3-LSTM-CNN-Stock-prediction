```python
import matplotlib.pyplot as plt

# Create a subplot of size 2x2
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot the "Close" column for the last 200 days
df[-200:].plot(y='Close', label='Close', ax=axs[0, 0])

# Calculate the EWMA for the last 200 days with span 1
ewma1 = df['Close'].ewm(span=1).mean()
ewma1[-200:].plot(label='EWMA (span=1)', ax=axs[0, 0])

# Add labels and legend to the first subplot
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Price')
axs[0, 0].legend()

# Plot the "Close" column for the last 200 days
df[-200:].plot(y='Close', label='Close', ax=axs[0, 1])

# Calculate the EWMA for the last 200 days with span 5
ewma5 = df['Close'].ewm(span=5).mean()
ewma5[-200:].plot(label='EWMA (span=5)', ax=axs[0, 1])

# Add labels and legend to the second subplot
axs[0, 1].set_xlabel('Date')
axs[0, 1].set_ylabel('Price')
axs[0, 1].legend()

# Plot the "Close" column for the last 200 days
df[-200:].plot(y='Close', label='Close', ax=axs[1, 0])

# Calculate the EWMA for the last 200 days with span 10
ewma10 = df['Close'].ewm(span=10).mean()
ewma10[-200:].plot(label='EWMA (span=10)', ax=axs[1, 0])

# Add labels and legend to the third subplot
axs[1, 0].set_xlabel('Date')
axs[1, 0].set_ylabel('Price')
axs[1, 0].legend()

# Plot the "Close" column for the last 200 days
df[-200:].plot(y='Close', label='Close', ax=axs[1, 1])

# Calculate the EWMA for the last 200 days with span 20
ewma20 = df['Close'].ewm(span=30).mean()
ewma20[-200:].plot(label='EWMA (span=20)', ax=axs[1, 1])

# Add labels and legend to the fourth subplot
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('Price')
axs[1, 1].legend()

# Adjust the layout of the subplots
plt.tight_layout()

# Show the plot
plt.show()

# Plot the "Close" column for the last 200 days
df[-200:].plot(y='Close', label='Close')

# Calculate the EWMA for the last 200 days
ewma = df['Close'].ewm(span=3).mean()

# Plot the EWMA for the last 200 days
ewma[-200:].plot(label='EWMA')

# Add labels and legend
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Show the plot
plt.show()
```
